```python
from pandas_datareader import data as web 
import numpy as np
import pandas as pd
from datetime import datetime as dt
from datetime import timedelta
import yfinance
import plotly.express as px

tickers = ['F','TM','GM','VWAGY','TSLA','NIO']

ticker_info = []

for t in tickers:
    print("Gathering info for ", t)
    a = yfinance.Ticker(t)
    ticker_info.append({'ticker':t,'info':a.info})   
```

```python
margin_tuples = []

for ti in ticker_info:
    margin_tuples.append((ti['ticker'],
                          ti['info']['grossMargins'],
                          ti['info']['ebitdaMargins'],
                          ti['info']['profitMargins']))
      

df_margins = pd.DataFrame(margin_tuples,columns=['Ticker','Gross Margins','EBITDA Margins','Profit Margins'])

fig = px.bar(df_margins,
             x=['Profit Margins','EBITDA Margins','Gross Margins'],
             y='Ticker',
             text_auto=True,
             barmode='group',
             height=1100)

fig.update_yaxes(title='Symbol', visible=True, showticklabels=True)
fig.update_xaxes(title='Margins', visible=True, showticklabels=True)
fig.show()
```

```python
fig = px.scatter(df_margins,
             x='Gross Margins',
             y='EBITDA Margins',
             hover_data=['Ticker'])

fig.update_yaxes(title='EBITDA Margins', visible=True, showticklabels=True)
fig.update_xaxes(title='Gross Margins', visible=True, showticklabels=True)
fig.show()
```
