# US Bond Data

```python
import pandas_datareader as pdr
import numpy as np
import matplotlib.pyplot as plt

df = pdr.get_data_fred(['GS2','GS5','GS10'])

df.head(10)
```

# Treasury Bond Rates

```python
df_dff = pdr.get_data_fred('DFF')

df.head(10)
```

```python
fig, ax = plt.subplots(figsize=(15,7))

plt.plot(df['GS2'].tail(24), label='2 Year US Treasury Bonds')
plt.plot(df['GS5'].tail(24), label='5 Year US Treasury Bonds')
plt.plot(df['GS10'].tail(24), label='10 Year US Treasure Bonds')
legend = ax.legend(loc='upper left', fontsize='x-large')
plt.title('2 Years of Treasurey Bonds Rates')
```

# Treasurey Bonds vs Federal Funds Rate

```python
fig, ax = plt.subplots(figsize=(15,7))

plt.plot(df['GS2'], label='2 Year US Treasury Bonds')
plt.plot(df_dff['DFF'], label='Federal Funds Effective Rate')
legend = ax.legend(loc='upper right', fontsize='x-large')
plt.title('2 Year Treasurey Bonds vs Federal Funds Effective Rate')
```

```python
fig, ax = plt.subplots(figsize=(15,7))

plt.plot(df['GS5'], label='5 Year US Treasury Bonds')
plt.plot(df_dff['DFF'], label='Federal Funds Effective Rate')
legend = ax.legend(loc='upper right', fontsize='x-large')
plt.title('5 Year Treasurey Bonds vs Federal Funds Effective Rate')
```
