# FRED CPI Data

```python
import pandas_datareader as pdr
import numpy as np
import matplotlib.pyplot as plt

df = pdr.get_data_fred('CPIAUCSL')

df.head(10)
```

# Percentage Change in CPI (12 Months)

```python
# get monthly inflation
df['PCT_CHANGE'] = df.pct_change(periods=12)

fig, ax = plt.subplots(figsize=(15,7))

plt.plot(df['PCT_CHANGE'],label='% Change in CPI from 12 Months Prior')
legend = ax.legend(loc='upper left', fontsize='x-large')

plt.title('Annual Change in CPI')
```
