```python
from pandas_datareader import data as web 
import numpy as np
import pandas as pd
from datetime import datetime as dt
from datetime import timedelta
import yfinance
import plotly.express as px

tickers = ['MSFT','SNOW','AAPL','TSLA','ASML','F','INTC','MU','NKE','MCO','RH','STNE','NIO','NU']

ticker_info = []

for t in tickers:
    print("Gathering info for ", t)
    a = yfinance.Ticker(t)
    ticker_info.append({'ticker':t,'info':a.info})


#print(ticker_info)
```

## Margins

```python
gross_margins = []

for ti in ticker_info:
    gross_margins.append(ti['info']['grossMargins'])
      
fig = px.bar(x=gross_margins,y=tickers,text_auto=True)
fig.update_yaxes(title='Symbol', visible=True, showticklabels=True)
fig.update_xaxes(title='Gross Margin', visible=True, showticklabels=True)
fig.show()
```

```python
ebitda_margins = []

for ti in ticker_info:
    ebitda_margins.append(ti['info']['ebitdaMargins'])
    
fig = px.bar(x=ebitda_margins,y=tickers,text_auto=True)
fig.update_yaxes(title='Symbol', visible=True, showticklabels=True)
fig.update_xaxes(title='EBITDA Margin', visible=True, showticklabels=True)
fig.show()
```

```python
profit_margins = []

for ti in ticker_info:
    profit_margins.append(ti['info']['profitMargins'])
    
fig = px.bar(x=profit_margins,y=tickers,text_auto=True)
fig.update_yaxes(title='Symbol', visible=True, showticklabels=True)
fig.update_xaxes(title='Profit Margin', visible=True, showticklabels=True)
fig.show()
```

```python
margin_tuples = list(zip(tickers,gross_margins,ebitda_margins,profit_margins))
df_margins = pd.DataFrame(margin_tuples,columns=['Ticker','Gross Margins','EBITDA Margins','Profit Margins'])
df_margins
```

```python
fig = px.bar(df_margins,
             x=['Profit Margins','EBITDA Margins','Gross Margins'],
             y='Ticker',
             text_auto=True,
             barmode='group',
             height=1100)

fig.update_yaxes(title='Symbol', visible=True, showticklabels=True)
fig.update_xaxes(title='Margins', visible=True, showticklabels=True)
fig.show()
```
